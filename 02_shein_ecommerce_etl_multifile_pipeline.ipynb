## 📌 Pandas Reference: `pd.to_numeric()`

### 1. What it does
* Converts an array, Series, or column of data into a proper numeric data type (either integers or floats).
* Essential for data pipelines because regular string functions (like `.str.extract()`) output text strings, not numbers you can do math with.

### 2. The Power of `errors='coerce'`
This parameter controls what happens when Pandas encounters a value it cannot turn into a number (e.g., text like `"unknown"`, `"N/A"`, or corrupt strings):
* **`coerce`:** Safely forces all non-convertible text into a true **`NaN`** (Not a Number) blank space instead of crashing your script.
* This allows you to chain a `.fillna(0)` right afterward to cleanly replace those errors with zeros.

```python
target_cols = ['rank-title', 'discount']

for col in target_cols:
    # 1. Extract digits and convert to numeric safely
    df_master[col] = pd.to_numeric(
        df_master[col].astype(str).str.extract(r'(\d+)', expand=False),
        errors='coerce'
    )
    # 2. Fill NaN with 0 and convert to integer
    df_master[col] = df_master[col].fillna(0).astype(int)

df_master.head(10)
```

## 📌 Code Summary: Missing Value Treatment & Type Casting

* Checks if `color-count` exists and fills its blank slots (`NaN`) with `0`. This keeps it a number so you can do math with it later.

### 2. Blanket Fill (`.fillna('unknown')`)
* Fills **every remaining empty spot** in the entire dataset with the word `'unknown'` in one single shot. This ensures there are no missing data gaps left.

### 3. Text Column Enforcement
* Loops through every column in the dataset.
* If the column isn't one of your numeric ones (`color-count`, `rank-title`, `discount`), it forces it into a clean string data type (`.astype(str)`). This standardizes all text fields (like titles and URLs).

```python
# 1. Fill the numeric column with 0
if 'color-count' in df_master.columns:
    df_master['color-count'] = df_master['color-count'].fillna(0)

# 2. Blanket fill EVERY other NaN in the entire dataframe with 'unknown' in one shot
df_master = df_master.fillna('unknown')

# 3. Force all text columns to be actual string data types
for col in df_master.columns:
    if col != 'color-count' and col != 'rank-title' and col != 'discount':
        df_master[col] = df_master[col].astype(str)

# 4. Print to verify the clean output
df_master.head(10)
```

## 📌 Code Summary: Targeted Missing Value Handling

### 1. Numeric Safe Guards
* Targets `color-count` first to fill its missing values (`NaN`) with `0`. This preserves its numeric data type so it doesn't get ruined by string conversions later.

### 2. Explicit Column Targeting
* Instead of doing a blanket fill on the whole dataset, you created a specific list (`text_cols`) containing only your known text and image link columns.

### 3. Isolated Text Cleaning
* Loops through your text list, checks if the column exists in the dataset, fills only its empty spots with `'unknown'`, and converts it to a clean string (`.astype(str)`).

### 4. Why this approach is better:
* It is much safer than a global blanket fill because it protects other columns (like `price` or `selling_proposition`) from accidentally being overwritten with the word `'unknown'` before they are properly cleaned!

```python
# 1. Handle the numeric column first so it stays a proper number
if 'color-count' in df_master.columns:
    df_master['color-count'] = df_master['color-count'].fillna(0)

# 2. The exact column names matching your table perfectly
text_cols = [
    'goods-title-link',
    'goods-title-link--jump',
    'goods-title-link--jump href',
    'rank-sub',
    'blackfridaybelts-bg src',
    'blackfridaybelts-content',
    'product-locatelabels-img src'
]

# 3. Fill missing gaps with 'unknown' while they are still recognized as NaN
for col in text_cols:
    if col in df_master.columns:
        df_master[col] = df_master[col].fillna('unknown').astype(str)

# 4. Print to verify the clean output
df_master.head(10)
```

## 📌 Code Summary: Custom Scale Function (`clean_sales`)

### 1. Pre-Cleaning & Text Standardization
* Forces incoming values to strings, lowers the case, and trims whitespace. If the text matches standard missing values (`'unknown'`, `'nan'`, etc.), it returns `0` immediately.

### 2. Smart Number Extraction
* Uses the regex pattern `([\d\.]+k?)` to grab only the digits, decimal points, and the optional letter `'k'` (e.g., extracting `"3.7k"` out of `"3.7k+ sold recently"`).

### 3. Multiplier Processing (`k` handling)
* Checks if `'k'` is present in the extracted string. If found, it strips the letter, converts the remaining text to a float, multiplies it by `1000`, and casts it to a clean integer (turning `"3.7k"` into `3700`). Otherwise, it casts standard raw digit strings directly into an integer.

### 4. Row-by-Row Mapping (`.apply()`)
* Uses `.apply(clean_sales)` to run this custom logic sequentially across every single row in the `selling_proposition` column, successfully sanitizing your sales volumes!

```python
def clean_sales(val):
    # Ensure it's a string and lowercase for easy matching
    val = str(val).lower().strip()

    # If it's a missing value or unknown, default it to 0
    if val in ['unknown', 'nan', 'none', '']:
        return 0

    # Extract just the numbers and periods/letters using a quick regex split
    # This catches things like '3.7k' or '700'
    match = pd.Series(val).str.extract(r'([\d\.]+k?)')[0].iloc[0]

    if pd.isna(match):
        return 0

    # Handle the 'k' multiplier
    if 'k' in match:
        num = match.replace('k', '')
        return int(float(num) * 1000)
    else:
        return int(float(match))

# Apply the cleaning function to the column
df_master['selling_proposition'] = df_master['selling_proposition'].apply(clean_sales)

# Print to see your perfectly calculated sales numbers!
df_master.head(10)
```

## 📌 Code Summary: Float Extraction & Safe Price Conversion

### 1. Robust Digit & Decimal Extraction
* Converts the `price` column to a string and applies the regex pattern `([\d\.]+)`.
* Unlike `\d+` (which breaks at periods), the `[\d\.]` character class captures both raw numbers and the literal decimal point, ensuring all cents remain perfectly intact (e.g., transforming `"$1.10"` to `"1.10"`).

### 2. Safe Data Type Conversion (`pd.to_numeric`)
* Converts the extracted string decimal directly into a float data type.
* Uses `errors='coerce'` as a failsafe; if a row contains unparseable text, it forces it into a true `NaN` blank instead of crashing the pipeline.

### 3. Missing Value Defaulting (`.fillna(0.0)`)
* Fills any missing or corrupted price values (`NaN`) with a default float value of `0.0`, ensuring data consistency across the entire column.

```python
df_master['price'] = pd.to_numeric(
    df_master['price'].astype(str).str.extract(r'([\d\.]+)', expand=False),
    errors='coerce'
)
df_master['price'] = df_master['price'].fillna(0.0).astype(float)
```

## Feature Engineering: Extracting Attributes from Titles

### 1. What it is
* Taking a raw string column (like `goods-title-link--jump` containing text like `"1pc Creative Toilet Sticker"`) and pulling out hidden data attributes into their own dedicated columns (like a `pack_size` column containing just `1`).

### 2. How it works (The Logic)
* You use a Regular Expression pattern like `r'^(\d+)(pc|pcs|pack|set)\b'` which scans the very beginning of the product title.
* It grabs the number (`\d+`) and checks if it's followed immediately by a keyword like `pc` (piece) or `pcs`.

### 3. Why this adds high value
* **Price Per Unit:** Knowing the pack size allows you to calculate the *true cost per piece* (e.g., Is a \$3 pack of 10 items cheaper per piece than a \$1 pack of 2 items?).
* **Filtering:** It lets you easily sort or group your 82,105 rows by single items versus multi-packs for deeper e-commerce business insights.

```python
df_master['pack_size'] = df_master['goods-title-link--jump'].str.extract(r'^(\d+)(?:pc|pcs|pack|set)', expand=False)
df_master['pack_size'] = df_master['pack_size'].fillna(1).astype(int)
df_master['pack_size']
```

### 1. Enables Granular Business Analytics
* Right now, your dataset has 82,105 rows mixed together. You can't easily answer questions like: *"What is the average discount for Hand Tools vs. Portable Fans?"*
* By isolating clean categories into a `sub_category` column, you can use `.groupby('sub_category')` to instantly compare market trends across different item classes.

### 2. Fixes Messy, Inconsistent Strings
* The raw data has structural inconsistencies because some rows start with `"in "` while others start with `"in Top rated in "`.
* Stripping these prefixes creates standard, clean text keys (e.g., changing `"in Hand Tools"` and `"in Top rated in Hand Tools"` both into just `"Hand Tools"`), preventing your groupings from breaking.

### 3. Powers Market Competitiveness Metrics
* Once you have clean categories combined with the **Original Price** calculation, you can determine which product niches on Shein rely on the most aggressive discount strategies to drive sales.

```python
# 1. Strip out the common ranking prefixes to isolate the clean category name
df_master['sub_category'] = df_master['rank-sub'].str.replace(r'^(in Top rated in |in )', '', regex=True).str.strip()

# 2. View the results to confirm "in Hand Tools" became "Hand Tools"
df_master[['rank-sub', 'sub_category']].head(10)
```

```python
df_master = df_master.copy()

df_master['original_price'] = np.where(
    df_master['discount'] < 100,
    df_master['price'] / (1 - (df_master['discount'] / 100)),
    df_master['price']
)

# 2. Round the new metric to 2 decimal places for clean currency formatting
df_master['original_price'] = df_master['original_price'].round(2)

# 3. View your new financial data structure
df_master[['price', 'discount', 'original_price']].head(10)
```
